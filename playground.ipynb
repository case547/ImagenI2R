```python
import torch
from models.img_transformations import DelayEmbedder
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

seq_len = 24

# (batch, length, features) -> (batch, features, embedding, embedding)
delay_embedder = DelayEmbedder(
    device,
    seq_len,
    delay=3,
    embedding=8,
    batch_size=1,
    num_features=1,
)

signal = torch.arange(1, seq_len+1).view(1, -1, 1).to(device)
signal.shape
```

```python
image_pad = delay_embedder.ts_to_img(signal)
image_no_pad = delay_embedder.ts_to_img(signal, pad=False)

print(image_pad, "\n", image_pad.shape, "\n")
print(image_no_pad, "\n", image_no_pad.shape)
```
